```python
import pandas as pd

df = pd.read_csv('galton_subset.csv')
df.head()
```

```python
df.plot.scatter(x="son", y="father")
```

```python
ax = df.plot.scatter(x="son", y="father", c="red", label='Father')
df.plot.scatter(x="son", y="mother", c="blue", label='Mother', ax=ax)

ax.set_xlabel("Son's height")
ax.set_ylabel("Parents' Height")
```

```python
import matplotlib.pyplot as plt

x = df['son']
y = df['father']
plt.scatter(x, y, c='red')
plt.show()
```

```python
son = df['son']
father = df['father']
mother = df['mother']

plt.scatter(son, father, c='red', label='Father')
plt.scatter(son, mother, c='blue', label='Mother')
plt.xlabel("Son's height")
plt.ylabel("Parents' height")
plt.legend()
plt.show()
```
