```python
import pandas as pd

my_data = {
    "Flavor": ['Chocolate', 'Strawberry', 'Vanilla'],
    "Number of Cartons": [16, 5, 9]
          }


df = pd.DataFrame(my_data)
df.head()
```

```python
df.plot.bar(x='Flavor', y='Number of Cartons', rot=0)
```

```python
df.plot.barh(x='Flavor', y='Number of Cartons', rot=0)
```

```python
df.plot.bar(x='Flavor', y='Number of Cartons', color="red", rot=0)
```

```python
import matplotlib.pyplot as plt

flavor = ['Chocolate', 'Strawberry', 'Vanilla']
number_of_cartons = [16, 5, 9]

plt.barh(flavor, number_of_cartons)
plt.xlabel("Flavor")
plt.ylabel("Number of Cartons")
plt.title("Icecream Statistics")

plt.show()
```
