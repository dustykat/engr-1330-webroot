```python
def computer_guess(num):
    low = 1
    high = 100
    guess = (low+high)//2
    while guess != num:
        guess = (low+high)//2
        print("The Computer Makes A Guess Of...", guess)
        if guess > num:
            high = guess
        elif guess < num:
            low = guess + 1

    print("The Computer Guesses", guess, "The Computer Is Right!")

def main():
    num = int(input("Enter A Number: "))
    if num < 1 or num > 100:
        print("Must Be In Range [1, 100]")
    else:
        computer_guess(num)

if __name__ == '__main__':
    main()
```

```python
def lagint(xlist,ylist,xpred): 
    lagint = 0.0 
    order = len(xlist)
    for i in range(order):
        term = ylist[i] 
        for j in range(order):
            if (i != j):
                term = term * (xpred-xlist[j])/(xlist[i]-xlist[j])
        lagint = lagint + term
    return(lagint)

xtable = [0.0,2.5,5.0,10,15.0,20.0]
ytable = [7.5,10.5,12.5,15.8,17.9,19.4]
xpred = [] 
ypred = [] 

size = 0.1 
many = int(1.0/size)


for i in range(many+1): 
    xpred.append(float(i)*size)
    ypred.append(lagint(xtable,ytable,float(i)*size))

    

import matplotlib.pyplot 
myfigure = matplotlib.pyplot.figure(figsize = (9,6))
matplotlib.pyplot.scatter(xtable, ytable ,color ='green') 
matplotlib.pyplot.plot(xpred, ypred, color ='blue') 
matplotlib.pyplot.xlabel("Mole Fraction - Liquid Phase") 
matplotlib.pyplot.ylabel("Mole Fraction - Vaopr Phase") 
mytitle = "Interpolating Polynomial Fit to Observations\n "
mytitle += "Green Markers are Observations " + "\n"
mytitle += "Blue Curve is Fitted Polynomial "+ "\n"
matplotlib.pyplot.title(mytitle) 
matplotlib.pyplot.show()
```

```python
def assignNewLabels(label):
    if label >= 7:
        return "Top"
    elif 5 < label < 7 :
        return "Average"
    elif label <= 5 :
        return "Low"
    else:
        return 'NA'
```

```python
winedata['QLabel'] = winedata['quality'].apply(assignNewLabels)
display(winedata)
```

```python
winedata['QLabel'].value_counts(normalize=True) * 100

lowwine = winedata.loc[(winedata['QLabel'] == 'Low')]
display(lowwine)

topwine = winedata.loc[(winedata['QLabel'] == "Top")]
display(topwine)
```

```python
import matplotlib.pyplot as plt
lowwine.plot(kind='hist',x='quality',y='pH',color='green', bins = 20)
plt.xlabel('pH Quality')
plt.ylabel('Samples')
plt.title('Low Quality Wine Samples and Its pH')
plt.show()
```

```python
topwine.plot(kind='hist',x='quality',y='pH',color='yellow', bins = 20)
plt.xlabel('pH Quality')
plt.ylabel('Samples')
plt.title('Top Quality Wine Samples and Its pH')
plt.show()

a = lowwine['pH']
b = topwine['pH']
plt.figure(figsize=(8,6))
plt.hist(a, bins=20, alpha=0.5, label="Low Quality", color = 'green')
plt.hist(b, bins=20, alpha=0.5, label="Top Quality", color = 'red')
plt.xlabel("pH Level", size=14)
plt.ylabel("Sample Frequency", size=10)
plt.title("Top Quality And Low Quality Wine pH")
plt.legend(loc='upper right')
```
