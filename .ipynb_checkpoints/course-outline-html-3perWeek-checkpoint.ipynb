# Typical Course Schedule (3 meetings/week)

- 3 lecture meetings/week (42 meetings)
- 3 lab meetings/week (42 meetings)
- Meetings 2 through 23 (week 1 through 8) are programming focused.
- Meetings 24 through 38 can be more data science focused to leverage programming skills
- Last three meetings are assigned non-assessed if using 3-exam model with a project; use last 3 meetings for project workshops/presentations
- By end of meeting 12 would have enough material to build a prediction engine to compute roots of numbers using a reduced instruction set (multiply, divide, add, subtract, compare, iterate) limitation as a computational thinking exercise
- By end of meeting 23 would have enough material to visually fit data models (curves) to observations (markers);
